```python
import pandas as pd
import numpy as np
import statistics as st
inah=pd.read_csv('INAH_detallado_2019.csv',header=0,  encoding='latin-1')
df=pd.DataFrame(inah)
print(df)
```

```python
df = df[["Estado", "Clave SIINAH", "Centro de trabajo", "Año", "Mes", "Tipo de visitantes", "Número de visitas", "Nacionalidad"]]
```

```python
print(df)
```

```python
df2 = df[["Estado", "Año", "Mes", "Número de visitas"]]
df3=df2.groupby(by=['Estado', 'Año', 'Mes']).count()
df3
```

```python
df_Estado_visitante = df[["Estado", "Tipo de visitantes", "Número de visitas"]]
df4=df_Estado_visitante.groupby(by=['Estado','Tipo de visitantes']).count()
df4
```

```python
dftipo_tiempo = df[["Estado", "Centro de trabajo", "Año", "Mes", "Tipo de visitantes", "Número de visitas"]]
df5=dftipo_tiempo.groupby(by=['Estado','Tipo de visitantes', 'Mes', 'Año']).sum()
df5
```
