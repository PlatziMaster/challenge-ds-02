1. Leer los datos del archivo **INAH_detallado_2019.csv (Presente en este repositorio)** utilizando un programa que de opción a insertarlo desde una interfaz gráfica simple o bien, desde línea de comandos. Pasando el archivo como un argumento.

```python
import pandas as pd
inah = pd.read_csv('INAH_detallado_2019.csv', encoding = 'ISO-8859-1')
inah
```

------------------------------------------------------------------------------------------------------------------------------
2. Limpiar los datos, estructurarlos usando python y pandas. Únicamente, utilizaremos: Estado, Clave SIINAH, Centro de trabajo, Año, Mes, Tipo de visitantes, Número de visitas y Nacionalidad.
------------------------------------------------------------------------------------------------------------------------------

```python
inah_original = inah.copy(deep=True)
```

```python
inah.drop(['Siglas'], axis=1, inplace=True)
```

```python
inah = inah.set_index(['Año', 'Mes', 'Estado']).sort_index()
inah
```

```python
inah['Número de visitas'] = inah['Número de visitas'].str.replace(',','')
inah['Número de visitas'] = pd.to_numeric(inah['Número de visitas'])
inah.dtypes
```

-------------------------------------------------------------------------------------------------------------------------------
3. Calculen los parámetros estadísticos:
-------------------------------------------------------------------------------------------------------------------------------

 - Totales por Estado por temporalidad (Mes y Año)

```python
totalsByStateAndTemporality = inah.sum(level=['Mes', 'Estado'])['Número de visitas']
totalsByStateAndTemporality
```

------------------------------------------------------------------------------------------------------------------------------ 
 - Totales por Estado por tipo de visitante

```python
visitors = pd.get_dummies(inah['Tipo de visitantes'])
visitors
```

```python
visitors = visitors.apply(lambda x: x * inah['Número de visitas'])
```

```python
totalVisitorsByState = visitors.sum(level=['Estado'])
totalVisitorsByState
```

-----------------------------------------------------------------------------------------------------------------------
- Totales por Estado por tipo de visitante y temporalidad (Mes y Año)

```python
totalVisitorsByStateAndTemporality = visitors.sum(level=['Mes', 'Estado'])
totalVisitorsByStateAndTemporality
```

----------------------------------------------------------------------------------------------------------------------------
- Totales por Centro de trabajo y temporalidad (Mes y año)

```python
totalByWorkCenterAndTemporality = inah.reset_index()
totalByWorkCenterAndTemporality = totalByWorkCenterAndTemporality.set_index(['Año', 'Mes', 'Centro de trabajo']).sort_index()
totalByWorkCenterAndTemporality.drop(['Estado','Clave SIINAH', 'Tipo de visitantes', 'Nacionalidad'], axis=1, inplace=True)
totalByWorkCenterAndTemporality
```

```python
totalByWorkCenterAndTemporality.sum(level=['Centro de trabajo','Año','Mes'])
```

----------------------------------------------------------------------------------------------------------------------------
- Totales por Centro de trabajo por tipo de visitante (Año)


```python
byWorkCenter = inah.reset_index()
byWorkCenter = byWorkCenter.set_index(['Año', 'Centro de trabajo']).sort_index()
```

```python
byWorkCenter
```

```python
visitorsType = pd.get_dummies(byWorkCenter['Tipo de visitantes'])
visitorsType
```

```python
pd.options.display.float_format = '{:.1f}'.format
visitorsType = visitorsType.apply(lambda x: x * byWorkCenter['Número de visitas'])
totalWorkCenterByvisitorsType = visitorsType.sum(level=['Año','Centro de trabajo'])
totalWorkCenterByvisitorsType
```

-----------------------------------------------------------------------------------------------------------
 - Promedio de visitantes totales por estado (Mes y Año)

```python
totalStates = inah.reset_index()['Estado'].nunique()
avVisitorsPerStatePerMonth = inah.sum(level=['Mes'])['Número de visitas'] / totalStates
avVisitorsPerStatePerMonth
```

-----------------------------------
 - Promedio de visitantes por tipo de visitante, por estado (Mes y Año)

```python
inah_data = inah.reset_index()
```

```python
totalVisitorsTypes = inah_data['Tipo de visitantes'].nunique()
```

```python
inah_data.groupby(['Año', 'Mes'])['Número de visitas'].sum().to_frame()/totalStates/totalVisitorsTypes
```

-----------------------
- Porcentaje de visitantes por Recinto en visitantes por estado (total por mes).

```python
percentageVisitors = inah_data.groupby(['Mes','Estado','Centro de trabajo'])['Número de visitas'].sum().to_frame()
percentageVisitors = percentageVisitors.groupby(level=[0,1]).apply(
    lambda x: x/x.sum() * 100
)
percentageVisitors.rename(columns={'Número de visitas':'Porcentaje visitas por estado'}, inplace=True)
percentageVisitors
```

----------------
- Porcentaje de tipo de visitantes por mes y año

```python
percentageVisitorsType = inah_data.groupby(['Año', 'Mes','Tipo de visitantes'])['Número de visitas'].sum().to_frame()
percentageVisitorsType = percentageVisitorsType.groupby(level=[0,1]).apply(
    lambda x: x/x.sum() * 100
)
percentageVisitorsType.rename(columns={'Número de visitas':'Porcentaje por tipo visitante'}, inplace=True)
percentageVisitorsType
```

--------------------------
- Porcentaje de tipo de visitantes por recinto (Mes y Año)

```python
percVisitorsPerworkCenter = inah_data.groupby(['Año', 'Mes','Centro de trabajo', 'Tipo de visitantes'])['Número de visitas'].sum().to_frame()
percVisitorsPerworkCenter = percVisitorsPerworkCenter.groupby(level=[0,1,2]).apply(
    lambda x: x/x.sum() * 100
)
percVisitorsPerworkCenter.rename(columns={'Número de visitas':'Porcentaje por tipo visitante por recinto'}, inplace=True)
percVisitorsPerworkCenter
```
